```python
# A custom implementation of matplotlib.pyplot with key methods

import matplotlib.pyplot as plt
import numpy as np

class CustomPyplot:
    def __init__(self):
        """Initialize the custom pyplot-like library."""
        print("Custom Pyplot Library Initialized!")

    def plot(self, x, y, label=None, color=None, linestyle=None):
        """Plot a line graph."""
        plt.plot(x, y, label=label, color=color, linestyle=linestyle)
        print("Line graph plotted.")

    def scatter(self, x, y, label=None, color=None, marker='o'):
        """Create a scatter plot."""
        plt.scatter(x, y, label=label, color=color, marker=marker)
        print("Scatter plot created.")

    def bar(self, x, height, label=None, color=None):
        """Create a bar chart."""
        plt.bar(x, height, label=label, color=color)
        print("Bar chart created.")

    def barh(self, y, width, label=None, color=None):
        """Create a horizontal bar chart."""
        plt.barh(y, width, label=label, color=color)
        print("Horizontal bar chart created.")

    def hist(self, data, bins=None, color=None, alpha=0.75, label=None):
        """Create a histogram."""
        plt.hist(data, bins=bins, color=color, alpha=alpha, label=label)
        print("Histogram created.")

    def pie(self, x, labels=None, autopct='%1.1f%%', colors=None):
        """Create a pie chart."""
        plt.pie(x, labels=labels, autopct=autopct, colors=colors)
        print("Pie chart created.")

    def imshow(self, img, cmap=None):
        """Display an image."""
        plt.imshow(img, cmap=cmap)
        print("Image displayed.")

    def contour(self, X, Y, Z, levels=None, cmap=None):
        """Create a contour plot."""
        plt.contour(X, Y, Z, levels=levels, cmap=cmap)
        print("Contour plot created.")

    def quiver(self, X, Y, U, V, color=None):
        """Plot vector fields."""
        plt.quiver(X, Y, U, V, color=color)
        print("Vector field plotted.")

    def fill_between(self, x, y1, y2=0, color=None, alpha=0.3):
        """Fill area between two curves."""
        plt.fill_between(x, y1, y2, color=color, alpha=alpha)
        print("Area between curves filled.")

    def boxplot(self, data, labels=None):
        """Create a boxplot."""
        plt.boxplot(data, labels=labels)
        print("Boxplot created.")

    def violinplot(self, data, showmeans=False):
        """Create a violin plot."""
        plt.violinplot(data, showmeans=showmeans)
        print("Violin plot created.")

    def title(self, title):
        """Set the title of the plot."""
        plt.title(title)
        print(f"Title set: {title}")

    def xlabel(self, label):
        """Set the x-axis label."""
        plt.xlabel(label)
        print(f"X-axis label set: {label}")

    def ylabel(self, label):
        """Set the y-axis label."""
        plt.ylabel(label)
        print(f"Y-axis label set: {label}")

    def legend(self):
        """Display the legend."""
        plt.legend()
        print("Legend displayed.")

    def grid(self, visible=True):
        """Toggle the grid."""
        plt.grid(visible)
        print(f"Grid {'enabled' if visible else 'disabled'}.")

    def savefig(self, filename):
        """Save the plot to a file."""
        plt.savefig(filename)
        print(f"Plot saved as {filename}.")

    def show(self):
        """Display the plot."""
        plt.show()
        print("Plot displayed.")

# Example usage
if __name__ == "__main__":
    # Initialize custom pyplot
    my_pyplot = CustomPyplot()

    # Example data
    x = np.linspace(0, 10, 100)
    y = np.sin(x)

    # Line plot
    my_pyplot.plot(x, y, label="Sine Wave", color="blue", linestyle="--")

    # Scatter plot
    my_pyplot.scatter(x, y + 0.5, label="Scatter Points", color="red")

    # Bar chart
    categories = ['A', 'B', 'C', 'D']
    values = [10, 20, 15, 25]
    my_pyplot.bar(categories, values, label="Bar Chart", color="green")

    # Titles, labels, and legend
    my_pyplot.title("Custom Pyplot Example")
    my_pyplot.xlabel("X-axis")
    my_pyplot.ylabel("Y-axis")
    my_pyplot.legend()

    # Show plot
    my_pyplot.show()
    
```
